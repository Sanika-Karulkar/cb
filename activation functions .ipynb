```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)

# Plot activation functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Function')
plt.grid(True)
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10, 100)

# Plot activation functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 2)
plt.plot(x, tanh(x))
plt.title('Tanh Function')
plt.grid(True)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-10, 10, 100)

# Plot activation functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 3)
plt.plot(x, relu(x))
plt.title('ReLU Function')
plt.grid(True)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # for numerical stability
    return exp_x / np.sum(exp_x, axis=0)
```

```python
x = np.linspace(-10, 10, 100)

# Plot activation functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 6)
softmax_values = softmax(x)
plt.plot(x, softmax_values)
plt.title('Softmax Function')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # for numerical stability
    return exp_x / np.sum(exp_x, axis=0)

# Generate x values
x = np.linspace(-10, 10, 100)

# Plot each activation function on a different graph
plt.figure(figsize=(12, 16))  # Adjust figsize to accommodate all subplots

plt.subplot(3, 2, 1)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Function')
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(x, tanh(x))
plt.title('Tanh Function')
plt.grid(True)

plt.subplot(3, 2, 3)
plt.plot(x, relu(x))
plt.title('ReLU Function')
plt.grid(True)

plt.subplot(3, 2, 4)  # Corrected index
softmax_values = softmax(x)
plt.plot(x, softmax_values)
plt.title('Softmax Function')
plt.grid(True)

plt.tight_layout()
plt.show()
```
